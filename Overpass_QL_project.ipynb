```python
#Using Overpass QL to query data from Open Street Map

import requests
import json
!pip install overpy
```

```python
#Finding the Name of Schools and universities in Nairobi
overpass_url = "http://overpass-api.de/api/interpreter"
overpass_query = """ 
  [out:json];
    area[name = 'Nairobi']->.searchArea;
    (
        way[building = 'school'](area.searchArea); 
        way[building = 'university'](area.searchArea);
    );
    out meta;
    out meta qt;
"""
```

```python
response = requests.get(overpass_url, params={'data': overpass_query})
data = response.json()
print(data['elements'])
```

```python
names = {}

for item in data['elements']:
  if 'name' in item['tags']:
    if not item['tags']['building'] in names:
      names[item['tags']['building']] = []
    if not item['tags']['name'] in names[item['tags']['building']]:
      names[item['tags']['building']].append(item['tags']['name'])
print(names)
```

```python
#Finding food eateries in Nairobi
overpass_query2 = """
  [out:json];
    area[name = 'Nairobi']->.searchArea;
    (
        node[amenity = 'restaurant'](area.searchArea);
        node[amenity = 'fast_food'](area.searchArea);
    );
    out meta;
    out meta qt;
"""
```

```python
response = requests.get(overpass_url, params={'data': overpass_query2})
data2 = response.json()
print(data2['elements'])
```

```python
restaurant_name = {}

for item in data2['elements']:
  if 'name' in item['tags']:
    if not item['tags']['amenity'] in restaurant_name:
      restaurant_name[item['tags']['amenity']] = []
    if not item['tags']['name'] in restaurant_name[item['tags']['amenity']]:
      restaurant_name[item['tags']['amenity']].append(item['tags']['name'])
print(restaurant_name)
```
